```python
#균등분포
def uniform_pdf(x):
    return 1 if x >= 0 and x < 1 else 0

def uniform_cdf(x):
    "returns the probability that a uniform random variable is less than x"
    if x < 0:   
        return 0    # uniform random is never less than 0
    elif x < 1: 
        return x    # e.g. P(X < 0.4) = 0.4
    else:       
        return 1    # uniform random is always less than 1

import numpy as np
x = np.arange(-1.0, 2.0, 0.1)

result_array = np.vectorize(uniform_pdf, otypes=[np.float])(x)

import matplotlib.pyplot as plt
%pylab inline

plt.plot(x, result_array)
plt.title("Uniform Distribution")
plt.axis([-1, 2, -1, 1.5])
plt.show()
```

```python
#표준정규분포
def normal_cdf(x, mu=0,sigma=1):
    return (1 + math.erf((x - mu) / math.sqrt(2) / sigma)) / 2

def plot_normal_cdfs(plt):
    xs = [x / 10.0 for x in range(-50, 50)]
    plt.plot(xs,[normal_cdf(x,sigma=1) for x in xs],'-',label='mu=0,sigma=1')
    plt.plot(xs,[normal_cdf(x,sigma=2) for x in xs],'--',label='mu=0,sigma=2')
    plt.plot(xs,[normal_cdf(x,sigma=0.5) for x in xs],':',label='mu=0,sigma=0.5')
    plt.plot(xs,[normal_cdf(x,mu=-1) for x in xs],'-.',label='mu=-1,sigma=1')
    plt.legend(loc=4) # bottom right
    plt.show()

import matplotlib.pyplot as plt
plot_normal_cdfs(plt)
```

```python
import numpy as np

min, max = 0, 1
x = np.linspace(min - .5, max + .5, 201)
print(x)
```

```python
y = []
for i in x:
    y.append((i >= min and i <=max) and (1 / (max - min)) or 0)
print(y)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12, 8))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Uniform Distribution')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def draw_z_score(x, cond, mu, sigma, title):
    y = norm.pdf(x, mu, sigma)
    z = x[cond]
    plt.plot(x, y)
    plt.fill_between(z, 0, norm.pdf(z, mu, sigma))
    plt.title(title)
    plt.show()

def draw_standard_normal_distribution(x, mu, sigma):
    y = norm.pdf(x, mu, sigma)
    plt.plot(x, y)
    plt.title("Standard normal distribution")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid()
    plt.show()
    
x = np.arange(-3,3,0.001)
draw_standard_normal_distribution(x, 0, 1)
```
