```
# 2 - 2
import numpy as np
onehots = np.array(['cat','the','dog','sat'])
onehots['cat']=[1,0,0,0]
onehots['the']=[0,1,0,0]
onehots['dog']=[0,0,1,0]
onehots['sat']=[0,0,0,1]
sentence = ['the','cat','sat']

def one_hot(array):
    unique, inverse = np.unique(array, return_inverse=True)
    onehot = np.eye(unique.shape[0])[inverse]
    return onehot

x = sentence
print("Sentence Encoding :" + str(x))
```

```
# 3 - 2
x = 3
y = -4
z = 2
w = 1

#forward pass
q = x * y 
p = z + w
f = q * p
print(q)
print(p)
print(f)

#backward pass
dfdp = q
dfdq = p
print(dfdp)
print(dfdp)

#p = z + w
dfdz = 1.0 * dfdp
dfdw = 1.0 * dfdp
print(dfdz)
print(dfdw)

#q = x * y 
dfdx = 1.0 * dfdq
dfdy = 1.0 * dfdq
print(dfdx)
print(dfdy)
```

```
# 4 - 1
import numpy as np
activation_function=lambda x: 2*x

Winput_output=np.matrix([[0.8,0.4],[0.9,0.6]])
print("Winput_output\n", Winput_output)

input = np.matrix([[1.0],[0.5]])
print("input.T\n", input)

output_bfunc=np.dot(Winput_output,input)
print("Xhidden\n", output_bfunc)

Output=activation_function(output_bfunc)
print("Ohidden\n", Output)
```

왜냐하면 인공지능이란 내가 어떠한 부분을 대신 해주기를 원해 아니면 이것을 해놔 라고 말하면은 인공지능이 이를 실행하는 것
즉 기계 및 컴퓨터가 더욱 좋은 능력을 가지고 있거나 아니면 기존의 컴퓨터가 하지못했던 이미지를 이해하거나 아니면 더욱 고차원 적인 분석을 내놓는 것이 아닌가 싶었기 때문입니다.
자세히 생각해보면 인공지능이란 결국 지능 즉 지금까지의 경험을 통해서 스스로 결론을 내는 능력을 가지고있는 인공적인 것을 말하는 것이다.
결국 우리의 명령 외에도 스스로 추가적인 작업을 할 수 있어야 한다는 것이다.
이러한 부분이 인공지능의 이해를 들었을때 가장 처음 아 내가 다른 생각을 가지고 있었구나 라고 느낀 부분이었습니다.
