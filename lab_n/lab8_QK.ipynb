```python
# 1 : 333 Neural network Forward propagation
import numpy as np

I = np.array([[0.9],[0.1],[0.8]])
Winput_hidden = np.array([[0.9,0.3,0.4],[0.2,0.8,0.2],[0.1,0.5,0.6]])
Whidden_output = np.array([[0.3,0.7,0.5],[0.6,0.5,0.2],[0.8,0.1,0.9]])

print(I)
print(Winput_hidden)
print(Whidden_output)

Xhidden = np.dot(Winput_hidden, I)
print(Xhidden)

def sigmoid(x):
    return 1 / (1 +np.exp(-x))

Ohidden = sigmoid(Xhidden)
print(Ohidden)

Xoutput = np.dot(Whidden_output, Ohidden)
print(Xoutput)

Ooutput = sigmoid(Xoutput)
print(Ooutput)
```

```python
# 2 : 222 Neural network Backward propagation
import numpy as np

error_output = np.array([[0.8],[0.5]])
Whidden_output = np.array([[2,1],[3,4]])
Whidden_input = np.array([[3,1],[2,7]])

print(error_output)
print(Whidden_output)

Wt_hidden_output = np.transpose(Whidden_output)
print(Wt_hidden_output)

Wt_hidden_input = np.transpose(Whidden_input)
print(Wt_hidden_input)

error_hidden = np.dot(Wt_hidden_output*1/2, error_output) # number of hidden nodes 2
print(error_hidden)

error_input = np.dot(Wt_hidden_input*1/2, error_hidden) # number of hidden nodes 2
print(error_input)
```

```python
# 3 : 222 Neural network Weight update
import numpy as np

def sigmoid(x):
    return 1 / (1 +np.exp(-x))

error_output = np.array([[0.8],[0.5]])
Whidden_output = np.array([[2,1],[3,4]])
Wt_hidden_output = np.transpose(Whidden_output)
O = np.array([[0.4],[0.5]])
update_Whidden_output = Whidden_output - ((-1)*error_output*sigmoid(np.dot(Wt_hidden_output, O))*
                                          (1-sigmoid(np.dot(Wt_hidden_output, O)))*O)
print(update_Whidden_output)
```

```python
# 5 : 333 Neural network Backward propagation & Weight update
import numpy as np

targets = np.matrix([[0.01], [0.01], [0.99]])
Whidden_output = np.array([[0.3,0.7,0.5],[0.6,0.5,0.2],[0.8,0.1,0.9]])
Winput_hidden = np.array([[0.9,0.3,0.4],[0.2,0.8,0.2],[0.1,0.5,0.6]])
Ooutput = np.array([[0.726],[0.708],[0.778]])
Input = np.array([[0.9],[0.1],[0.8]])

print(targets, "targets\n")
print(Whidden_output, "Whidden_output\n")
print(Winput_hidden, "Winput_hidden\n")
print(Input, "Input\n")


def sigmoid(x):
    return 1 / (1 +np.exp(-x))

Wt_hidden_output = np.transpose(Whidden_output)
print(Wt_hidden_output, "Wt_hidden_output\n")

Wt_input_hidden = np.transpose(Winput_hidden)
print(Wt_input_hidden, "Wt_input_hidden\n")

error_output = targets - Ooutput
print(error_output, "error_output\n")

error_output_T = np.transpose(error_output)
print(error_output_T, "error_output_T\n")

Ohidden = np.array([[0.761],[0.603],[0.65]])
print(Ohidden, "Ohidden\n")

Ohidden_T = np.transpose(Ohidden)
print(Ohidden_T, "Ohidden_T\n")

Input_T = np.transpose(Input)
print(Input_T, "Input_T\n")

#출력 계층 은닉계층 사이 가중치 업데이트 및 역전파
error_hidden = np.dot(Wt_hidden_output*1/3, error_output) # number of hidden nodes 3
print(error_hidden, "error_hidden\n")

print(sigmoid(np.dot(Wt_hidden_output, Ohidden)), "sigmoid sigma Wjk Oj\n")

update_Whidden_output = Whidden_output - ((-1)*error_output*np.transpose(sigmoid(np.dot(Wt_hidden_output, Ohidden)))*
                                          (1-sigmoid(np.dot(Wt_hidden_output, Ohidden)))*Ohidden_T)
print(update_Whidden_output, "update_Whidden_output\n")

#은닉 계층과 입력 계층 사이 
error_input = np.dot(Wt_hidden_input*1/3, error_hidden) # number of hidden nodes 3
print(error_input,"error_input\n")

print(sigmoid(np.dot(Wt_hidden_input, Input)), "sigmoid sigma\n")

update_Winput_hidden = Winput_hidden - ((-1)*error_hidden*np.transpose(sigmoid(np.dot(Wt_input_hidden, Input)))*
                                          (1-sigmoid(np.dot(Wt_input_hidden, Input)))*Input_T)
print(update_Winput_hidden, "update_Winput_hidden\n")
```
