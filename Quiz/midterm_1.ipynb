```
# 문제 4 - 2
import numpy as np
def L1_loss(output,target): 
    return np.abs(output-target)

activation_function=lambda x: 2*x

target = np.array([[1.2],[1.9]])
print("target\n", target)

output = np.array([[2.0],[2.4]])
print("output\n", output)

input = np.array([[1.0],[0.5]])
print("input\n",input)

error_output_notL1 = target - output
print("error_output_notL1\n", error_output_notL1)

error_output = L1_loss(output,target)
print("error_output\n", error_output)

Winput_output = np.array([[0.8,0.4],[0.9,0.6]])
print("Winput_output\n",Winput_output)

Winput_output_sum1=Winput_output.sum(axis=1, dtype='float')
print("Winput_output_sum1\n", Winput_output_sum1)
Winput_output_norm=Winput_output.T/Winput_output_sum1.T
print("Winput_output_norm\n", Winput_output_norm)

error_input=np.dot(Winput_output_norm, error_output)
print("error_input\n", error_input)
```

```
# 문제 4 - 3
import numpy as np
#활성화함수
activation_function=lambda x: 2*x

# L1(절대값) 손실
def L1_loss(output,target): 
    return np.abs(output-target)

input = np.array([[1.0],[0.5]]) # 입력값들
print("input\n",input)

Winput_output = np.array([[0.8,0.4],[0.9,0.6]]) #입력-출력 노드 사이의 가중치
print("Winput_output\n",Winput_output)

Xoutput=np.dot(Winput_output,input) # 활성화 함수 적용전 결과값
print("Xoutput\n", Xoutput)

output=activation_function(Xoutput) #결과값 or 실제 결과값
print("output\n", output)

target = np.array([[1.2],[1.9]])  #예측 결과값
print("target\n", target)

error_output_notL1 = target - output  #출력 오차
print("error_output_notL1\n", error_output_notL1)

error_output = L1_loss(output,target) # L1 손실 적용 출력 오차
print("error_output\n", error_output)

Winput_output_sum1=Winput_output.sum(axis=1, dtype='float') # 정규화를 위한 가중치 값 덧셈
print("Winput_output_sum1\n", Winput_output_sum1)
Winput_output_norm=Winput_output.T/Winput_output_sum1.T # 정규화를 적용한 가중치 값들
print("Winput_output_norm\n", Winput_output_norm)

error_input=np.dot(Winput_output_norm, error_output)  #입력 오류
print("error_input\n", error_input)

# 가중치 업데이트 과정
WjkOj=np.dot(Winput_output, input)
print("WjkOj\n", WjkOj)

S=activation_function(WjkOj)
print("S\n", S)

delEoverdelWjk = -np.dot(error_output*S*(1-S), input.T)
print("delEoverdelWjk\n", delEoverdelWjk)

Winput_output = Winput_output - (0.1 * delEoverdelWjk)
print("new Winput_output\n", Winput_output)
```

1. "인공지능", "머신러닝", "딥러닝"의 차이점이 무엇이라고 생각하시는지요?

A : 인공지능은 머신러닝과 딥러닝을 포함하고 있는 가장 큰 개념입니다.
인공지능이란 기계 및 컴퓨터를 이용하여 인공적으로 만든 지능을 말합니다.
더 이야기 하자면 기계 혹은 컴퓨터가 인간의 행동 또는 생각 처리 등을 모방하는 것을 말합니다.
예를들면 바둑을 두는 프로그램인 알파고 혹은 스타크래프트 게임을 했던 알파스타와 같은 것들을 의미합니다.

머신러닝은 인공지능의 하위 개념이며 딥러닝을 포함하는 개념입니다.
머신러닝이란 기계학습이라는 말로 컴퓨터 혹은 기계가 테이터를 학습을 하는 것을 말합니다.
데이터를 학습하면서 데이터 안에서 유의미한 결과, 동일성 규칙성 등을 발견 및 추출 하여 이를 활용하는 것을 말합니다.

딥러닝은 머신러닝의 하위개념입니다.
딥러닝은 심층학습이라는 말로 입력과 출력 사이에 은닉계층을 여러개 넣은 인공신경망을 연구하는 것을 말합니다.
딥러닝은 더욱 복잡한 문제를 해결할 수 있도록 기계학습을 실행하기 위한 알고리즘 혹은 기술들을 말합니다.

2. 과제로 KMOOC "인공지능의 이해" 강좌를 청강하셧습니다. 여러분이 "인공지능의 이해" 강좌에서 가장 중요하게 습득한 개념, 이론, 또는 지식은 무엇이었는지요?

A : 인공지능의 개발의 가장 중요한 목표는 시키지 않는 것을 하는것 스스로 판단하는것 이 부분이 저에게 다르게 다가온 부분이었습니다. 
왜냐하면 인공지능이란 내가 어떠한 부분을 대신 해주기를 원해 아니면 이것을 해놔 라고 말하면은 인공지능이 이를 실행하는 것 즉 기계 및 컴퓨터가 더욱 좋은 능력을 가지고 있거나 아니면 기존의 컴퓨터가 하지못했던 이미지를 이해하거나 아니면 더욱 고차원 적인 분석을 내놓는 것이 아닌가 싶었기 때문입니다. 
자세히 생각해보면 인공지능이란 결국 지능 즉 지금까지의 경험을 통해서 스스로 결론을 내는 능력을 가지고 있어야 합니다.
그러므로 사람의 명령 외에도 스스로 추가적인 작업을 할 수 있어야 한다는 것이다. 이러한 부분이 인공지능의 이해를 들었을때 가장 처음 아! 내가 다른 생각을 가지고 있었구나 라고 느낀 부분이었습니다.
